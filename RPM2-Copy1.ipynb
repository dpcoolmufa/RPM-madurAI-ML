```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.svm import LinearSVC
from sklearn.multiclass import OneVsOneClassifier
from sklearn import cross_validation
```

```python
df = pd.read_excel("students_dataSet.xls")
print (df)
```

```python
from sklearn.model_selection import train_test_split
X=df.iloc[0:2100,[0,1,2,5,7,8,9,10,25]]
print(X)
y=df.iloc[0:2100,[0,1,2,5,7,8,9,10]]
print(y)
X=np.array(X)
y=np.array(y)
```

```python
print(X.shape,"\t",y.shape)
```

```python
from sklearn import svm
classifier=svm.SVC(kernel='linear',C=1.0)
classifier.fit(X,y)
classifier.predict(X)
```

```python
x_train, x_test, y_train, y_test = cross_validation.train_test_split(X,y,test_size = 0.2,random_state=5)
y_test_pred = classifier.predict(x_test)
```

```python
predictions=[round(i) for i in y_test_pred]
accuracy=classifier.score(y_test_pred.reshape(-1,1),predictions)
```

```python
print("Accuracy: %.2f%%" % ( accuracy * 100.0))
```
